```python
import math
import numpy as np
import pandas as pd
from datetime import datetime, timedelta, date
import os


chunksize = 10000
number_of_areas = 265
size_timestep = 60
number_of_timesteps = int(12 * 60 / size_timestep)

array_matrix = np.zeros(number_of_timesteps * number_of_areas**2).reshape(number_of_timesteps,number_of_areas,number_of_areas)
array_sum = np.zeros(number_of_timesteps * number_of_areas**2).reshape(number_of_timesteps,number_of_areas,number_of_areas)
array_counter = np.zeros(number_of_timesteps * number_of_areas**2).reshape(number_of_timesteps,number_of_areas,number_of_areas)

if (os.path.isfile('test2.txt')):
    os.remove('test2.txt')

# loop for all start-to-end-areas (should they be saved?)
# do it the other way around
a = 0

for chunk in pd.read_csv('2018_5_add_duration.csv', chunksize = chunksize):
    #? oder laufen lassen
    counter = 0
    while(counter < chunksize):
        try:
            area1 = chunk['PULocationID'][a] - 1
            area2 = chunk['DOLocationID'][a] - 1

            #Check timestep
            string_pickup = chunk['tpep_pickup_datetime'][a]
            help_time_pickup = (datetime.strptime(string_pickup,'%m/%d/%Y %I:%M:%S %p') - timedelta(hours = 6)).time()
            #print(help_time_pickup)
            help_timeslot = (datetime.combine(date.min, help_time_pickup) - datetime.min ) / 60
            timeslot = math.floor(help_timeslot.seconds / 60)
            #print(help_timeslot)
            #print(timeslot)

            #take this out after manipulation data again
            if(timeslot == number_of_timesteps):
                timeslot = number_of_timesteps -1

            array_sum[timeslot,area1,area2] += chunk['trip_distance'][a]
            array_counter[timeslot,area1,area2] += 1
        except:
            pass
        a += 1
        counter += 1
    print(a)

# Get average    
d = 0
while (d < number_of_timesteps):     
    b = 0
    while (b < number_of_areas):
        c = 0
        while(c < number_of_areas):
            # print(d,b,c)
            if (array_counter[d,b,c] != 0):
                array_matrix[d,b,c] = array_sum[d,b,c] / array_counter[d,b,c]
            else:
                array_matrix[d,b,c] = 0
            c += 1    
        b += 1 
    d += 1
    print(d,b,c)

#save in .txt file

with open('test2.txt', 'w') as outfile:

    outfile.write('# Array shape: {0}\n'.format(array_matrix.shape))

    for data_slice in array_matrix:
        np.savetxt(outfile, data_slice, fmt='%-4.2f') 

        outfile.write('# New slice\n')
```
