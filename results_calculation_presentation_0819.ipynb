```python
# results melgarejo
import re

filename = 'results_noTW.txt'
x = 0
sum_10 = 0
sum_20 = 0
sum_30 = 0
sum_50 = 0
sum_100 = 0 

with open(filename) as f:
    for line in f:
        if('matrix10' not in line and 'matrix20' not in line):
            #print(line)
            #print(line[:7])
            if(line[:7] == 'inst_10' and line[:8] != 'inst_100'):
                numbers = re.findall(r"\d+",line)
                rightnumber = [int(e) for e in numbers if int(e) > 1000]
                sum_10 += rightnumber[0]

            if(line[:7] == 'inst_20'):
                numbers = re.findall(r"\d+",line)
                rightnumber = [int(e) for e in numbers if int(e) > 1000]
                sum_20 += rightnumber[0]

            if(line[:7] == 'inst_30'):
                numbers = re.findall(r"\d+",line)
                rightnumber = [int(e) for e in numbers if int(e) > 1000]
                sum_30 += rightnumber[0]

            if(line[:7] == 'inst_50'):
                numbers = re.findall(r"\d+",line)
                rightnumber = [int(e) for e in numbers if int(e) > 1000]
                sum_50 += rightnumber[0]

            if(line[:8] == 'inst_100'):
                numbers = re.findall(r"\d+",line)
                rightnumber = [int(e) for e in numbers if int(e) > 1000]
                sum_100 += rightnumber[0]
            
print(sum_10,sum_20,sum_30,sum_50,sum_100)
```
