```python
# with time-depedencies with average
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
import os


chunksize = 100
i = 1
j = 1
number_of_areas = 265
size_timestep = 60
number_of_timesteps = int(12 * 60 / size_timestep)
array = np.zeros(number_of_timesteps * number_of_areas**2).reshape(number_of_timesteps,number_of_areas,number_of_areas)

if (os.path.isfile('test2.txt')):
    os.remove('test2.txt')

# loop for all start-to-end-area-trips
while (i <= number_of_areas):
    while (j <= number_of_areas):
        a = 0
        summe = 0
        counter = 0
        average = 0
        for chunk in pd.read_csv('short_tripdata.csv', chunksize = chunksize):
            while(a < chunksize):
                if (chunk['PULocationID'][a] == i and chunk['DOLocationID'][a] == j):
                    
                    # !!! Are they in the right timeslot
                    found = False
                    b = 0
                    lower_bound = datetime.strptime('12/07/1990 06:00:00 AM','%m/%d/%Y %I:%M:%S %p').time()
                    upper_bound = lowerbound + timedelta(minutes = size_timestep)
                    
                    while(not found and b < number_of_timesteps):
                        string_pickup = chunk['tpep_pickup_datetime'][a]
                        help_time_start = datetime.strptime(string_pickup,'%m/%d/%Y %I:%M:%S %p').time()
                        
                        if(lower_bund < help_time_start and help_time_start < upper_bound):
                                                                        
                            summe += chunk['trip_distance'][a]
                            counter += 1
                            found = True
                            
                        lower_bound += timedelta(minutes = size_timestep)
                        upper_bound += timedelta(minutes = size_timestep)
                        b += 1
            
                a += 1
                #print(i , j , a)
        
        # write in array
        
        if(counter != 0):
            array[b,i-1,j-1] = summe / counter
            print(i-1,j-1)        
        
        j += 1
    print(i , j , a)
    i += 1
    j = 0


with open('test2.txt', 'w') as outfile:

    outfile.write('# Array shape: {0}\n'.format(array.shape))

    for data_slice in array:
        np.savetxt(outfile, array, fmt='%-4.2f') 

        outfile.write('# New slice\n')
```
